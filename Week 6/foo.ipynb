```python
data_num = data[['Id', 'Age', 'DistanceFromHome', 'Education', 'EmployeeNumber', 'EnvironmentSatisfaction', 'JobInvolvement', 'JobSatisfaction', 'MonthlyIncome', 'NumCompaniesWorked', 'PercentSalaryHike', 'PerformanceRating', 'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager', 'CommunicationSkill', 'Behaviour']].copy()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.figure(figsize=(10, 10), dpi=100)
sns.heatmap(data_num.corr())
```

```python
data_uc = data_num[['Age','DistanceFromHome','EnvironmentSatisfaction', 'JobInvolvement','JobSatisfaction','StockOptionLevel','TrainingTimesLastYear']].copy()

data_uc.head()
```

```python
data_cat = data[['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'OverTime']].copy()

data_cat.head()
```

```python
data_cat = pd.get_dummies(data_cat)

data_cat.head()
```

```python
data_final = pd.concat([data_num, data_cat], axis=1)
data_final.head()
```

```python
# from sklearn.model_selection import train_test_split

y_train = data_final.Attrition
X_train = data_final.drop('Attrition', axis=1)

# X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.4, random_state=10)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()

model.fit(X_train, y_train)
```
